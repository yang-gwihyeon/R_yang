```r

의사결정나무 <- function(iris_data){
    library(caret)
    library(party)
    
# Sampling
samp <- createDataPartition(iris_data$대기시간, p=0.7, list=F)
# 학습용 데이터 만들기
data.tr <- iris_data[samp,]

# 검증용 데이터 만들기
data.test <- iris_data[-samp,]

model.ctree <- ctree(대기시간 ~ ., data=data.tr)

x <- subset(data.test, select = -대기시간)
y <- data.test$대기시간

pred <- predict(model.ctree, x, type = "class") # class :분류된것(객관식), raw : 주관식


 mean(pred == y)
    }
```

```r

의사결정나무 <- function(iris_data){
    library(caret)
    library(party)
    
# Sampling
samp <- createDataPartition(iris_data$Species, p=0.7, list=F)
# 학습용 데이터 만들기
data.tr <- iris_data[samp,]

# 검증용 데이터 만들기
data.test <- iris_data[-samp,]

model.ctree <- ctree(Species ~ ., data=data.tr)

x <- subset(data.test, select = -Species)
y <- data.test$Species

pred <- predict(model.ctree, x, type = "class") # class :분류된것(객관식), raw : 주관식


 mean(pred == y)
    }
```
