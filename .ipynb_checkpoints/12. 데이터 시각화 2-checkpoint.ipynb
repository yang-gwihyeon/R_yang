## 선 굵기 변경

```r
x <- c(1,3,5,2,9,10)
plot(x,
    type = "b",
    col = c("red","blue"),
    lwd = 3 # 선굵기
    )
```

```r
### 한 화면에 차트 여러개 그려서 비교하기

price <- runif(10, 1, 100)
plot(price)
```

```r
# 2행 2열 차트 그리기 : type 속성으로 구분해보기

par(mfrow = c(2,2))

plot(price, type="l")
plot(price, type="o")
plot(price, type="h")
plot(price, type="s")
```

```r
# 2행 2열 차트 그리기 : pch 속성으로 구분해보기
```

```r
par(mfrow = c(2,2))
plot(price, type="o", pch=5)
plot(price, type="o", pch=15)
plot(price, type="o", pch=25)
plot(price, type="o", pch=c(rep(15,8), 20, 20), 
                      col="blue",
                        cex = c(rep(1,8),3,1))
```

### 중첩자료 시각화

```r
# Method 1)
xy.df <- as.data.frame(table(x, y))
xy.df
plot(x, y,
    pch= 15,
    cex = 2 * xy.df$Freq
    )
```

```r
options(warn = -1)
par(family = "AppleGothic")

#기타 그래프
x <- c( 1, 3, 6, 8, 9)
y <- c(12,56,78,32, 9)

plot(x, y)

# 선분 그리기
segments(6, 78, 8, 32, col="red")

# 화살표 그리기
arrows(3,56,1,12,col="blue")

# 사각형 그리기
rect(4, 20, 6, 30, col="red", border = "red", density = 10)

#글자 쓰기
text(4, 40, "This is a 샘플", srt=50) #str 대각선

#테두리 활용하기
mtext("character", side=3, adj=0)

# chart의 테두리 그리스
box(lty=1, col= "red")

# x축의 추가 : y축의 40위치에 0~10까지 표기하고 색은 빨간색
axis(1, pos=40, at=0:10, col="red")

#y축의 추가 : x축의 6dnlcldp 20~60까지 표시하고 색은 파란색
axis(2, pos=6, at=20:60, col='blue')
```

```r
par(family="AppleGothic")
options(warn=-1)
options(scipen=100)

a <- t(arrange(election,desc(pop))[1:5,5:7])
cc <- t(arrange(election,desc(pop))[1:5,4:7][1])
a[1]
cc
#colnames(a) <- c("경기도 화성시","경기도 부천시","경기도 남양주시","서울특별시 송파구","서울특별시 강서구")
barplot(a,
        col= c("sky blue","red","light green"),
        main = "투표수 높은 지역의 후보별 투표수",
        ylim = c(0,500000),
        names = paste(arrange(election,desc(pop))[1:5,2],arrange(election,desc(pop))[1:5,3]),
        cex.names = 0.7,
       )

legend(5, 468538,
      c("안철수","홍준표","문재인"),
      col=c("light green","red","sky blue"),
      pch=c(15,15,15),

      )

tt <- 0.7
ttt <- 0

for (w in 1:5){
    for (i in 1:3){
      text(tt, 
           a[i,w]/2+ttt, 
         paste(round(a[i,w]/sum(arrange(election,desc(pop))[w,5:7]),4)*100,"%"))
    ttt = ttt + a[i,w]
}
    tt = tt + 1.2
    ttt = 0
}
```

```r
d<-arrange(election,desc(pop))[,c(3,4,5,6,7)]
d$tot <- d$moon+d$hong+d$ahn

d$문<-round((d$moon/d$tot)*100,2)
d$홍<-round((d$hong/d$tot)*100,2)
d$안<-round((d$ahn/d$tot)*100,2)
head(d,5)

for(i in 1:5){
text(0.7+1.2*(i-1),d$moon[i]/2,paste(d$문[i],"%"))
text(0.7+1.2*(i-1),d$hong[i]/2+d$moon[i],paste(d$홍[i],"%"))
text(0.7+1.2*(i-1),d$ahn[i]/2+d$moon[i]+d$hong[i],paste(d$안[i],"%"))
}
```
