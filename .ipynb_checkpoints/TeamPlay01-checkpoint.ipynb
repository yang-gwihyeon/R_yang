```r
seoul <- read.csv("../R_yang/newSeoul2019.csv")
```

```r
library(dplyr)
```

```r
str(seoul)
```

```r

seoulMonths <- data.frame(월=seoul$월,발생건수=seoul$발생건수,부상자수=seoul$부상자수,사망자수=seoul$사망자수)
head(seoulMonths)
```

```r
#arrange(seoulMonths,(월))
seoulMonth1 <- subset(seoulMonths,월 == 1)
#seoulJanuary <- seoulMonths[]
```

```r
seoulMonthz <- data.frame(month=1,발생건수=sum(seoulMonth1$발생건수),부상자수=sum(seoulMonth1$부상자수),사망자수=sum(seoulMonth1$사망자수))
```

```r
seoulMonthz
```

```r


for(i in 2:12){
    seoulMonth1 <- subset(seoulMonths,월 == i)
    seoulMonthz <- rbind(seoulMonthz,data.frame(month=i,발생건수=sum(seoulMonth1$발생건수),부상자수=sum(seoulMonth1$부상자수),사망자수=sum(seoulMonth1$사망자수)))
}
```

```r
seoulMonthz
```

```r
options(scipen = 100)
options(warn = -1)
par(family = "AppleGothic")
barplot(seoulMonthz$발생건수, # GDP상위 15개국 Data
        main = "월별 발생건수",
        xlab = "월",
        ylab = "단위(1건)",
        names = seoulMonthz$month,
        cex.names = 0.7,
        col = c('grey','yellow',rep('grey',7),'red','grey'),
        ylim = c(0, 40000)
     
       )
```

```r
options(scipen = 100)
options(warn = -1)
par(family = "AppleGothic")
barplot(seoulMonthz$부상자수, # GDP상위 15개국 Data
        main = "월별 부상자수",
        xlab = "월",
        ylab = "단위(1명)",
        names = seoulMonthz$month,
        cex.names = 0.7,
        col = c('grey','yellow',rep('grey',7),'red','grey'),
        ylim = c(0, 50000)
     
       )
```

```r
options(scipen = 100)
options(warn = -1)
par(family = "AppleGothic")
barplot(seoulMonthz$사망자, # GDP상위 15개국 Data
        main = "월별 사망자",
        xlab = "월",
        ylab = "단위(1명)",
        names = seoulMonthz$month,
        cex.names = 0.7,
        col = c('grey','yellow',rep('grey',6),'red','grey','grey','grey'),
        ylim = c(0, 300)
     
       )
```

```r
par(family = "AppleGothic")
plot(seoulMonthz$발생건수,seoulMonthz$부상자수)
cor(seoulMonthz$발생건수, seoulMonthz$부상자수)
```

```r
par(family = "AppleGothic")
plot(seoulMonthz$사망자,seoulMonthz$부상자수)
cor(seoulMonthz$사망자, seoulMonthz$부상자수)
```

```r
par(family = "AppleGothic")
plot(seoulMonthz$발생건수,seoulMonthz$사망자)
cor(seoulMonthz$부상자수, seoulMonthz$사망자)
```
