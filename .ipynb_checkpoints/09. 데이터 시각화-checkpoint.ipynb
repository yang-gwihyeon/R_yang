# 선그래프 

```r
# 3종류의 dataset의 선그래프
abc <- c(260, 300, 250, 280, 310)
def <- c(180, 200, 210, 190, 170)
ghi <- c(210, 250, 260, 210, 270)
```

```r
par(family = 'AppleGothic')
plot(abc,
    type = "o",
    col = "red",
    ylim = c(0, 400),
    axes = F, # x y좌표 Frame 제거
    ann = F # x,y의 label 제거
    )

    # x축과 y축의 라벨을 설정
    axis(1, at=1:5, lab = c('서울','대전','대구','광주','원주'))

    # y축과 y축의 라벨을 설정
    axis(2, ylim = c(0, 400))

    # main title 지정
    title(main = "과일 판매량", col.main = "red")

    # x축 title(label) 지정
    title(xlab = "지역", col.lab = "blue")

    # y축 title(label) 지정
    title(ylab = "판매량", col.lab = "blue")

    # 선그리기 함수로 def 그리기
    lines(def,
        type = "o",
          pch = 22,
          col = "green",
          lty = 2
         )

    lines(ghi,
        type = "o",
          pch = 22,
          col = "blue",
          lty = 2
         )

    #범례
legend(4, 400, # x, y좌표
      c("야구장", "축구장", "해변가"),
        col = c("red", "green", "blue"),
       pch= c(21, 22, 22),
       lty = c(1, 2, 2),
       cex = 0.8
      )
```

# 막대 그래프

```r
x <- 1:6
barplot(x,
       main = "1~6",
        xlab = "Sales",
        ylab = "Amounts",
        names = c("A","B","C","D","E","F"),
        col = c(rep("red",3), rep("blue",3))
       )
```

```r
# 복잡한 형태의 막대 그래프
# 3행 2열의 matrix 생성

x <- matrix(1:6, 3, 2)
x 
barplot(x,
       ylim = c(0, 16)
       )
```

```r
# 누적 막대 그래프 분리 후 열별 Label 붙이기
par(family = "AppleGothic")

rownames(x) <- c('1월','2월','3월')
barplot(x,
        beside = T, # 누적부분 해체
        ylim = c(0, 8),
        names = c("korea", "America"),
#        names = c("k1","k2","k3","A1","A2","A3"),
        main = "수출량",
        xlab = "국가",
        ylab = "빈도수",
        col = c("red","green","blue"),
        legend.text = T
       )

```

# 학력별 점수 차이

```r
data <- read.csv("../R_yang/education.csv", header = F)
```

```r
str(data)
```

```r
summary(data)
```

```r
#V1 : 남, 여
#V2 : 점수
#V3 : 국, 영, 수
#V4 : 초, 중, 고, 대, 원

data$V1 <- factor(data$V1, levels = 1:2, labels = c('남','여'))
data$V3 <- factor(data$V3, levels = 1:3, labels = c('국어','영어','수학'))
data$V4 <- factor(data$V3, levels = 1:5, labels = c('초','중','고','대','원'))
```

```r
str(data)
```

```r
# 컬림 이름 변경 (성별, 점수, 과목, 학력)
colnames(data) = c("성별","점수","과목","학력")
```

```r
summary(data)
```

```r
table(data$점수)
```

```r
par(family = 'AppleGothic')
barplot(table(data$점수),
       main = "점수별 빈도수",
       xlab = "점수",
        ylab = "빈도수",
        ylim = c(0,12)
        
       )
```

```r
# 성별 학력별 빈도수를 표현
table(data$성별)
table(data$학력)
table(data$성별, data$학력)
table(data$학력, data$성별)
```
