```r
완성잠실 <- subset(완성,관광특구=="잠실")
완성잠실 <- arrange(완성잠실,desc(방문외국인수))
완성잠실
```

```r
visit명동<-subset(완성, 관광특구 == "잠실")
visit명동 <- arrange(visit명동, 년도)

par(family="AppleGothic",bg='white'); options(warn=-1)
options(scipen=100)
plot(visit명동$방문외국인수,(visit명동$점포당매출금액*0.001)
    , main = '잠실 - 방문외국인수 대비 점포당매출금액'
    , ylim = c(0,160000)
    , xlim = c(0,100000)
    , ylab = "점포당매출금액 (단위:천원)"
    , xlab = "방문외국인수 (단위:명)"
   
    )
r <-lm((visit명동$점포당매출금액*0.001)~visit명동$방문외국인수)
abline(r, col='red')
cor(visit명동$방문외국인수,visit명동$점포당매출금액)
```
