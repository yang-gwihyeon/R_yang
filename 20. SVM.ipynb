```r
# 표준화 : 평균을 0으로 표준편차는 1로 변경
head(iris)
iris2 <- subset(iris, select = -Species)
iris3 <- scale(iris2)
summary(iris3)
```

```r
iris4 <- as.data.frame(iris3)
```

```r
iris4$Species <- iris$Species
head(iris4)
```

```r
library(randomForest)
library(caret)
library(party)
```

```r
# Sampling
samp <- createDataPartition(iris4$Species, p=0.7, list=F)
# 학습용 데이터 만들기
data.tr <- iris4[samp,]

# 검증용 데이터 만들기
data.test <- iris4[-samp,]

#model.ctree <- ctree(Species ~ ., data=data.tr)



# 문제 및 정답 만들기
x <- subset(data.test, select = -Species)
y <- data.test$Species
```

```r
# 학습시키기
model.svm = svm(Species ~ ., data=data.tr)
```

```r
# 검증하기
pred <- predict(model.svm, x)
pred
```

```r
# 예측율 및 오분류표
mean(pred == y)
table(pred, y)
```
