# 선그래프 

```r
# 3종류의 dataset의 선그래프
abc <- c(260, 300, 250, 280, 310)
def <- c(180, 200, 210, 190, 170)
ghi <- c(210, 250, 260, 210, 270)
```

```r
par(family = 'AppleGothic')
plot(abc,
    type = "o",
    col = "red",
    ylim = c(0, 400),
    axes = F, # x y좌표 Frame 제거
    ann = F # x,y의 label 제거
    )

    # x축과 y축의 라벨을 설정
    axis(1, at=1:5, lab = c('서울','대전','대구','광주','원주'))

    # y축과 y축의 라벨을 설정
    axis(2, ylim = c(0, 400))

    # main title 지정
    title(main = "과일 판매량", col.main = "red")

    # x축 title(label) 지정
    title(xlab = "지역", col.lab = "blue")

    # y축 title(label) 지정
    title(ylab = "판매량", col.lab = "blue")

    # 선그리기 함수로 def 그리기
    lines(def,
        type = "o",
          pch = 22,
          col = "green",
          lty = 2
         )

    lines(ghi,
        type = "o",
          pch = 22,
          col = "blue",
          lty = 2
         )

    #범례
legend(4, 400, # x, y좌표
      c("야구장", "축구장", "해변가"),
        col = c("red", "green", "blue"),
       pch= c(21, 22, 22),
       lty = c(1, 2, 2),
       cex = 0.8
      )
```

# 파이차트


```r
sales <- c(210, 110, 400, 550, 600, 130)
pie(sales)
```

```r
par(family = "AppleGothic")
sales <- c(210, 110, 400, 550, 600, 130)
week <- c("월","화","수","목","금","토")
ratio <- sales / sum(sales) * 100
label <- paste(week,"\n",ratio,"%")
```

```r
par(family = 'AppleGothic')
pie(sales,
    main = "영업실적 파이차트",
    labels = label,
    col = rainbow(length(sales))

   )
```

# 3D Pie Chart

```r
library(plotrix)
```

```r
sales <- c(210, 110, 400, 550, 600, 130)
week <- c("월","화","수","목","금","토")
ratio <- sales / sum(sales) * 100
label <- paste(week,"\n",ratio,"%")

par(family = 'AppleGothic')
pie3D(sales,
    main = "영업실적 파이차트",
    labels = label,
    explode = 0.05,
    labelcex = 0.8,
    radius = 1
   )
```
