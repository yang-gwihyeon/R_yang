```r
### 서울시 교통사고 데이터로 발생건수와 부상자 수와 산포도 구하기
seoul <- read.csv("../R_yang/newSeoul2019.csv")
head(seoul)
```

```r
#발생건수의 숫자가 150이상이면 빨간색 아니면 파란색 표현하기

par(family="AppleGothic")
plot(seoul$발생건수, seoul$부상자수,
    main = "서울시 교통사고 발생건수와 부상자수 산포도",
    xlab ="발생건수",
    ylab ="부상자수",
    col = ifelse(seoul$발생건수 >=150 , "red", "blue")
    )
```

```r
# main title : Simple Time Series
y <- c(1,3,2,4,8,5,7,6,9)
```

```r
#발생건수의 숫자가 150이상이면 빨간색 아니면 파란색 표현하기

par(family="AppleGothic")
plot(y,
    main = "Simple Time Series",
    xlim = c(min(y)-1, max(y)+1),
    ylim = c(min(y)-1, max(y)+1),
    )
```
