```r

의사결정나무 <- function(iris_data){
    library(caret)
    library(party)
    
# Sampling
samp <- createDataPartition(iris_data$등급, p=0.7, list=F)
# 학습용 데이터 만들기
data.tr <- iris_data[samp,]

# 검증용 데이터 만들기
data.test <- iris_data[-samp,]

model.ctree <- ctree(등급 ~ ., data=data.tr)

x <- subset(data.test, select = -등급)
y <- data.test$등급

pred <- predict(model.ctree, x, type = "class") # class :분류된것(객관식), raw : 주관식


 mean(pred == y)
    }
```

```r
# 인공 신경망 모델 사용

인공신경망 <- function(iris_data){
    

library(nnet)
library(caret)

# Sampling
samp <- createDataPartition(iris_data$등급, p=0.7, list=F)
# 학습용 데이터 만들기
data.tr <- iris_data[samp,]

# 검증용 데이터 만들기
data.test <- iris_data[-samp,]

# Model 만들기
model.nnet <- nnet(등급 ~ ., data=data.tr, size=4) 
# size:은닉층 갯수로 sample수보다 작아야 함(Hyper Parameter)

x <- subset(data.test, select = -등급)
y <- data.test$등급

pred <- predict(model.nnet, x, type = "class") # class :분류된것(객관식), raw : 주관식


 mean(pred == y)
    
}
```
