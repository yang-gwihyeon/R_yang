```r
x <- 10
y <- runif(1) * 100

if(x>y){
    print("x>runif()")
}
print("---End---")
```

```r
x <- 10
y <- runif(1) * 100

if(x>y){
    print("x>runif()")
}else{
    print("x<=runif()")
}
print("---End---")
```

```r
if(x>y){
    print("x>runif()")
}else if (x<=y){
    print("x<=runif()")
}else {
    print("x = runif()")
}
print("---End---")
```

```r
# excel에서 q3 컬럼 데이터의 홀수의 합과 짝수의 합 구하기

odd <- 0
even <- 0
for(i in excel$q3){

    if(i%%2==1){
          odd = odd+i
    }else{
        even = even+i
    }
}

cat("홀수의 합은",odd,"이고 짝수의 합은", even,"입니다.")


# 홀수의 합은 604 이고 짝수의 합은 852 입니다.
```

```r
three <- 0
for(i in excel$q3){
    if(i%%3==0){
        three = three + i
    }
}
cat("3의 배수의 합은",three,"입니다")
```

```r
excel$q3
thour <- 0
for(i in excel$q3){
    if(i%%3==0 && i%%4==0){
        thour = thour + i
    }
}
cat("3,4의 배수의 합은",thour,"입니다")
```

```r

for(i in 1:ncol(test)){
    
    cat(i,"번째 컬럼 빈도수의 최대값과 최소값","\n")
    cat("최대값:",max(table(test[,i])),", 최소값 :",min(table(test[,i])),"\n")
    cat("----------------------","\n")
}
```

```r
# 각 컬럼의 빈도수 구하기와 각 칼럼의 빈도수와 최대값,최소값 구하기
tableMaxMin <- function(x){
    
    for(i in 1:ncol(x)){
    
    cat(colnames(x)[i],"컬럼의 빈도 분석 결과","\n")
    
    print(table(x[,i]))
    cat("======================","\n")
   
    
    cat(i,"번째 컬럼 빈도수의 최대값과 최소값","\n")
    cat("최대값:",max(table(x[,i])),", 최소값 :",min(table(x[,i])),"\n")
    cat("----------------------","\n")
}

    
}
```
